```python
n_max = 1000

primes = []

for x in range(n_max+1):
    if sympy.isprime(x):
        primes.append(x)
        
primes
```

Quite clunky <br>
List comprehension becomes useful

```python
primes = [x for x in range(n_max + 1) if sympy.isprime(x)]
# reads like english
# Do not do onliner for the sake of it, keep in mind readability

primes
```

```python
# 2018 nobel prize
names = (
    "Frances Arnold",
    "George Smith",
    "Gregory Winter",
    "postponed",
    "Denis Mukwege",
    "Nadia Murad",
    "Arthur Ashkin",
    "Gérard Mourou",
    "Donna Strickland",
    "James Allison",
    "Tasuku Honjo",
    "William Nordhaus",
    "Paul Romer",
)

nationalities = (
    "USA",
    "USA",
    "UK",
    "---",
    "DRC",
    "Iraq",
    "USA",
    "France",
    "Canada",
    "USA",
    "Japan",
    "USA",
    "USA",
)

categories = (
    "Chemistry",
    "Chemistry",
    "Chemistry",
    "Literature",
    "Peace",
    "Peace",
    "Physics",
    "Physics",
    "Physics",
    "Physiology or Medicine",
    "Physiology or Medicine",
    "Economics",
    "Economics",
)
```

```python
# We want a list of tuple of each category
[(cat, name, nat) for name, nat, cat in zip(names, nationalities, categories)]
```

```python
# Code like poetry easier to read
[
    (cat, name, nat) 
    for name, nat, cat in zip(names, nationalities, categories)
]
```

```python
# We want only for the chemistry category
[
    (cat, name, nat) 
    for name, nat, cat in zip(names, nationalities, categories)
    if cat == 'Chemistry'
]
```

```python
chem_names = [
    name
    for name, cat in zip(names, categories)
    if cat == 'Chemistry'
]

chem_names
```

```python
# all possible pairs of chemistry laureates
[
    (n1, n2)
    for i, n1 in enumerate(chem_names)
    for j, n2 in enumerate(chem_names)
    if i > j
]

# Can have nested loop
# Here condition if to avoid having 2 times the same pair
```

```python
# If we want a else statement
# Wrong to do the following (it does not generalize)

[
    ("Medecine", name, nat) 
    for name, nat, cat in zip(names, nationalities, categories)
    if cat == "Physiology or Medicine" else (cat, name, nat) 
    
]
```

```python
# Instead do
[
    ("Medecine", name, nat) if cat == "Physiology or Medicine" else (cat, name, nat) 
    for name, nat, cat in zip(names, nationalities, categories) 
]
```

```python
[
    (cat, name, nat) 
    for name, nat, cat in zip(names, nationalities, categories)
]
```

## Dictionary comprehension

```python
{
    name: (cat, nat) 
    for name, nat, cat in zip(names, nationalities, categories)
}
```

# High level plotting with iqplot

```python
import numpy as np
import pandas as pd

import iqplot

import bokeh.io

bokeh.io.output_notebook()
```

```python
df = pd.read_csv('data/frog_tongue_adhesion.csv', comment='#')

# Have a look so we remember
df.head()
```

```python
# quite declarative
p = iqplot.box(
    data=df,
    q='impact force (mN)',
    cats='ID'
)

bokeh.io.show(p)

# By defaut quantitative data in the x axis
```

```python
# Plot all data:
p = iqplot.strip(
    data=df,
    q='impact force (mN)',
    cats='ID'
)

bokeh.io.show(p)

# Notice frogs I uniform distribution,
# frogs III very consistent
```

```python
# Plot all data:
p = iqplot.stripbox(
    data=df,
    q='impact force (mN)',
    cats='ID'
)

bokeh.io.show(p)

# Great overlay annotation
```

```python
p = iqplot.histogram(
    data=df,
    q='impact force (mN)',
    cats='ID'
)

bokeh.io.show(p)
```

```python
p = iqplot.histogram(
    data=df,
    q='impact force (mN)',
    cats='ID',
    density = True, #area under curve = 1
)

bokeh.io.show(p)
```

```python
p = iqplot.ecdf(
    data=df,
    q='impact force (mN)',
    cats='ID',
)

bokeh.io.show(p)
```

```python
p = iqplot.ecdf(
    data=df,
    q='impact force (mN)',
    cats='ID',
    style='staircase' #each corner is a measured data, so better way to represent than histogram, look at slop
)

bokeh.io.show(p)
```
