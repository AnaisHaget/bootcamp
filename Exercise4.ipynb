## Exercise 4.3: Working with two-dimensional arrays

```python
A = np.array(
    [
        [6.7, 1.3, 0.6, 0.7],
        [0.1, 5.5, 0.4, 2.4],
        [1.1, 0.8, 4.5, 1.7],
        [0.0, 1.5, 3.4, 7.5],
    ]
)

b = np.array([1.1, 2.3, 3.3, 3.9])
```

```python
print(A)
print(b)
```

```python
print('Row 1 of A: ', A[1,:])
print(' ')
print('Col 1 and 3 of A: ', A[:,[1,3]])
print(' ')
print('A values greater than 2:', A[A>2])
print(' ')
print('Diagonal of A:', np.diag(A))
```

```python
x = np.linalg.solve(A, b)
print(x, '\n')
print(b, '\n')
print(np.dot(A, x), '\n')
print(np.transpose(A), '\n')
```

```python
B = np.ravel(A)
print(A)
print(B)
```

## Exercise 4.4: Understanding and building ECDFs

```python
import numpy as np
import pandas as pd

import bokeh.io
import bokeh.plotting

bokeh.io.output_notebook()
```

```python
# Dummy data set as a Pandas Series
rg = np.random.default_rng()
data = pd.Series(rg.normal(0, 1, size=100))

# Compute y-values for ECDF
ecdf_y = data.rank(method='first') / len(data)

# Make the plot
p = bokeh.plotting.figure(
    frame_height=200,
    frame_width=300,
    x_axis_label='x',
    y_axis_label='ECDF',
)

p.circle(data, ecdf_y)

bokeh.io.show(p)
```

```python
def ecdf_vals(data):
    """takes a one-dimensional NumPy array of data and returns the x and y values for plotting a “dot-style” ECDF.
    That is, each dot has a y value given by the ECDF evaluated at x."""
    
    x = np.sort(data)
    # numpy.arange([start, ]stop, [step, ]dtype=None, *, like=None)
    y = np.arange(1, data.size + 1) / data.size
    
    return x, y
```

```python
A=np.array([1, 1, 0, 1, 1, 1, 1, 1, 1])
x, y = ecdf_vals(A)
```
