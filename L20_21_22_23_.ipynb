```python
mu = 10
sigma = 1

x = rg.normal(mu, sigma, 100000)

p = iqplot.histogram(
    x,
    density=True,
    rug=False,
    bins=100
)

bokeh.io.show(p)
```

```python
mu = 10
sigma = 1

x = rg.normal(mu, sigma, 100000)

p = iqplot.ecdf(
    x[:1000], # we do not want too much data points
)

bokeh.io.show(p)
```

```python
# Overly theorethical plot
mu = 10
sigma = 1

x = rg.normal(mu, sigma, 100000)

x_theor = np.linspace(6, 14, 400)
y_theor = scipy.stats.norm.cdf(x_theor, mu, sigma)

p = iqplot.ecdf(
    x[:5000], 
)

p.line(
    x=x_theor,
    y=y_theor,
    line_width=2,
    line_color="tomato",
)

bokeh.io.show(p)
```
