# Exercise 4.1: Using Git/GitHub to submit this homework

This may be your first time using Git/GitHub. [Git](https://git-scm.com) is a version control system that allows for collaborative work. [GitHub](http://github.com/) is a website that can host Git repositories. We will be using Git/GitHub exclusively for submitting the homework exercises. We do this for two main reasons. First, this allows for much easier tracking, submission, and grading of homework. Second, and most importantly, you should be using version control when working with real data. It is good practice both for the security of your own work and for reproducibility of your research.

We have created a repository hosted on GitHub, hosted at `https://github.com/broadinbroad/2021_fsri_exercises`, that you will be using to submit your solutions to the exercises and that we will be using to provide feedback. **This provides directions for submitting via command line, but you can also use** [GitHub Desktop](https://desktop.github.com/) **if you prefer.**

**a)** Check that you are a collaborator on the `2021_fsri_exercises` repository. Heidi needs to add you as a collaborator, so email or Slack her with your GitHub username. **If you skip this step, you will not be able to submit your homework.**

**b)** Clone your repository to your machine.I like to keep my repositories in the directory `~/git`, but the location on your machine is your choice. 

To clone into the `git` directory, I would do the following on my machine on the command line. (You can get a command prompt by launching a Terminal using the JupyterLab launcher.) **Note that `~` denotes the directory where** `git` **is located. Replace this with whatever the path is to your** `git` **folder on your local machine.**

    cd ~/git
    git clone https://github.com/broadinbroad/2021_fsri_exercises.git

When you clone, you may be asked for your username and password. You can just type your username. For the password, you may need a public access token (see directions [here](https://docs.github.com/en/github/authenticating-to-github/keeping-your-account-and-data-secure/creating-a-personal-access-token)). You can copy and paste that token as your password; for security, it will not appear on your screen.

**c)** Upon cloning the repository, you will have a new directory, `~/git/2021_fsri_exercises` containing the exercise submission repository. Within that directory you will see three subdirectories, `week_1/`, `week_2/`, and `week_3/`, where you will be submitting your solutions to the problems from each respective week of the bootcamp. 

Each week, you will need to create a personal directory (i.e. `lastname/`) in that week's folder.  We will start with `week_1/`.  In your local machine, navigate to the `week_1` folder within the repository. For my machine, I would write in the terminal:

    mkdir ~/git/2021_fsri_exercises/week_1/miller

Within this folder (e.g. `week_1/miller`), we will need to make two more directories, one called `data/` and one called `sandbox`.  So from inside your named folder, do:

    mkdir ~/git/2021_fsri_exercises/week_1/miller/data
    mkdir ~/git/2021_fsri_exercises/week_1/miller/sandbox

These two directories will *not* be under version control, and anything you put in there will *not* be uploaded to GitHub. Git will automatically ignore the contents of these directories because it is included in the `.gitignore` file of the repository that we created for you.

What are these two directories for? The `data/` folder is where you will place the datasets you download for any problems that that require them. We are having Git ignore this folder because we do not want to have large data files under version control. The `sandbox/` directory is where you will place your solution files as you work on them before they are ready to be submitted.  This is because many people will be pulling from and pushing to the repo as they submit their work, and you do not want to accidentally overwrite progress you've made on your work with an older version of the file when you pull from the repo to your local computer before pushing your submission.  If you work on things in the `sandbox/` before you're ready to submit them, you can be sure they'll be safe on your local computer.


**d)** You will place the solution to this homework problem in the file `lastname_exercise4.1.ipynb` (with your own last name of course).  Start out by creating the file in the `sandbox/` folder.  In a Markdown cell in the Jupyter notebook, write a haiku (an unrhymed three-line poem/verse, where the first line has 5 syllables, the second 7 syllables, and the third line 5 syllables) about something you're looking forward to in your first year as a Caltech student. (Don't put too much time or thought into the haiku; it's just for fun to add some text.) 

**e)** When you are ready to submit the notebook, first check to see if any changes have been made to the repo.  You can do this by typing the following into the command line:

    git status

If you're told that `Your branch is up to date with 'origin/main'.`, then you're good to proceed to the steps below this one.  If changes have been made to the repo, then pull the updates first by typing:

    git pull

Now save your exercise/haiku file and move it out of the sandbox and into your named folder above it (e.g. `miller/`).  Then perform the command line prompts written below (but with your last name, of course) to add it, commit, and push.

    cd ~/git/2021_fsri_exercises/week_1/miller
    git add miller_exercise4.1.ipynb
    git commit -m "Commit of Exercise 4.1 with haiku."
    git push origin main
    
As you are working through your notebooks, it is wise to commit and push often. This is especially important if you collaborate with others on work in the future.  In cases where you may be working with teammates, before you start working, you will want to pull to get any changes made to the repo. Additionally, to avoid merge conflicts with teammates, you may want to edit notebooks separately and then copy and paste them into the notebooks you will submit at the end. These other notebooks should be in the `sandbox/` directory to keep your version controlled directory clean. More experienced Git users can use branches and merges instead.
